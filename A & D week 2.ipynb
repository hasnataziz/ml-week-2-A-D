# 2. Data Cleaning:## 
o Handle missing values by filling them with the median or mode, or by usin 
other appropriate imputation methods.

# filling the missing value in age by median

```python
train["Age"].fillna(train["Age"].median())
```

```python
train.isnull().sum()
```

# now its time to fill the missing value of cabin

```python
train["Cabin"] = train["Cabin"].fillna("unknown")
```

```python
train.isnull().sum()
```

### Now there are no missing values in the required column which is given in the problem

# Remove outliers if necessary or treat them appropriately.
as we see up in the boxplot there are outliers in fare column so here is the code below to remove outliers
### this can be done by IQR method

```python
q1 = train["Fare"].quantile(0.25)
q3 = train["Fare"].quantile(0.75)
iqr = q3-q1
iqr
```

```python
lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr
```

```python
train = train[(train['Fare'] >= lower_bound) & (train['Fare'] <= upper_bound)]
train
```
